```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('energydata_complete.csv')
df.describe()
```

```python
scaler = MinMaxScaler()
drop = df.drop(columns=['date', 'lights'])
normalised_df = pd.DataFrame(scaler.fit_transform(drop), columns=drop.columns)
features_df = normalised_df.drop(columns=['Appliances'])
target = normalised_df['Appliances']
```

```python
features_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_df, target,
test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)
predicted_values
```

# Question 13
### What is the Mean Absolute Error (in two decimal places)?

```python
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

# Question 14
### What is the Residual Sum of Squares (in two decimal places)?

```python
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

# Question 15
### What is the Root Mean Squared Error (in three decimal places)?

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

# Question 16
### What is the Coefficient of Determination (in two decimal places)? 

```python
r2_score_1 = r2_score(y_test, predicted_values)
round(r2_score_1, 2)
```
