```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
url= 'https://github.com/WalePhenomenon/climate_change/blob/master/fuel_ferc1.csv?raw=true'
fuel_data = pd.read_csv(url, error_bad_lines=False)
fuel_data.describe(include='all')
```

```python
#checking the mean for the average of each numerical data
fuel_data.groupby('fuel_type_code_pudl').mean()

#checking the correlation between the "fuel cost per unit burned" and other numerical data
# "fuel cost per unit delivered" for nuclear type and "measure of energy per unit" is higher in relation
```

```python
fuel_data.groupby('fuel_qty_burned')['report_year'].count()
```

```python
#check for missing values
fuel_data.isnull().sum()
```

```python
#checking the statsistics for the highest fuel cost per unit delivered grouped by year
fuel_data.groupby('report_year')['fuel_cost_per_unit_delivered'].describe()
```

```python
by_fuel_type = fuel_data.groupby(['fuel_type_code_pudl','report_year'])
```

```python
coal_mean94 = by_fuel_type.get_group(('coal', 1994))['fuel_cost_per_unit_delivered'].mean()
coal_mean98 = by_fuel_type.get_group(('coal', 1998))['fuel_cost_per_unit_delivered'].mean()

gas_mean94 = by_fuel_type.get_group(('gas', 1994))['fuel_cost_per_unit_delivered'].mean()
gas_mean98 = by_fuel_type.get_group(('gas', 1998))['fuel_cost_per_unit_delivered'].mean()
```

```python
#percentage drop in delivery of coal btw 1998 and 1994
coal_mean94/coal_mean98 *100
```
